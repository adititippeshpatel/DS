## The line of code helps in finding patterns in the data by analyzing correlations between numerical features of the Titanic dataset. This can reveal potential relationships between different factors that may influence the survival rate, ticket fare, etc., such as:

Whether passengers who paid more for tickets (fare) were older (age).

Whether people traveling with siblings/spouses (sibsp) were also traveling with parents/children (parch).

Pearson correlation quantifies the linear relationship between two continuous variables. It produces a correlation matrix, where:

Values close to 1 indicate a strong positive linear relationship (as one variable increases, the other also increases).

Values close to -1 indicate a strong negative linear relationship (as one variable increases, the other decreases).

Values close to 0 indicate no linear relationship between the variables.


```python
titanic_cleaned.select_dtypes(include='number').corr(method='pearson')
```

## If the histogram is skewed to the right, it would indicate that most passengers paid lower fares, with a few passengers paying very high fares.

If the histogram is roughly uniform, it suggests that the fare distribution was more even across all passengers.

If there is a sharp peak, it could indicate a large number of passengers were in a particular fare range.

```python
import seaborn as sns
import matplotlib.pyplot as plt
# Plot histogram for 'fare'
plt.figure(figsize=(8, 5))
plt.hist(titanic['fare'], bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of Ticket Fare')
plt.xlabel('Fare')
plt.ylabel('Number of Passengers')
plt.grid(True)
plt.show()
```

## This line plots both a histogram and a KDE curve to visualize the distribution of ticket fare prices in the Titanic dataset.
Kernel Density Estimate (KDE) is a way to estimate the probability distribution of a dataset, providing a smooth curve over the data
a histogram can look "blocky" because it breaks the data into bins. KDE tries to give you a smoother, continuous curve that shows how likely it is to find a data point at any specific value.

```python
# Distplot (deprecated in Seaborn 0.11+)
sns.distplot(titanic['fare'], bins=30, color='green')
plt.title('Fare Distribution using distplot')
plt.xlabel('Fare')
plt.ylabel('Density')
plt.show()
```

## Use distplot() (only if you're using an older version of Seaborn < 0.11):

If you need both histogram and KDE curve together in one line.

Use histplot() (if you're using Seaborn 0.11+ or later):

For more flexibility in creating histograms, adjusting the KDE curve, rug plots, and the overall visual appearance.

It's the recommended function for statistical visualization of distributions going forward.



```python
# Histplot
sns.histplot(data=titanic, x='fare', bins=30, kde=True, color='purple')
plt.title('Fare Distribution using histplot')
plt.xlabel('Fare')
plt.ylabel('Count')
plt.show()
```

## A JointPlot is a type of plot that helps us explore the relationship between two continuous variables (numeric features) in a dataset. It combines a scatter plot with marginal histograms (or density plots) to show both:

How the two variables are related (via the scatter plot).

The distribution of each variable individually (via the marginal histograms).



```python
# Jointplot: fare vs age
sns.jointplot(data=titanic, x='fare', y='age', kind='scatter', color='orange')
plt.suptitle('Jointplot of Fare vs Age', y=1.02)
plt.show()
```

## A rug plot is a simple way to visualize one-dimensional distributions. It places small ticks or "rug marks" along an axis at the data points. It's a minimalistic and efficient way of displaying the distribution of a single variable.

If the ticks are closely packed at certain points, it means many passengers paid similar fares.

If the ticks are spread out, it indicates diverse fares, with passengers paying a variety of ticket prices.

```python
# Rugplot for 'fare'
plt.figure(figsize=(8, 1))  # Flat layout since rugplots are 1D
sns.rugplot(x=titanic['fare'], height=0.5, color='teal')
plt.title('Rug Plot of Fare')
plt.xlabel('Fare')
plt.show()
```

## It shows the average fare paid by passengers in each class (First, Second, Third).
