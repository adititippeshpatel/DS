## We want to understand how a numeric value (like income or age) changes across different categories (like age groups).

In your dataset:

Categorical Variable: Age Group → values like "Teen", "Middle Age", "Elder"

Numeric Variable: Annual Income (k$)
df.groupby('Age Group'):

Groups all rows into 3 categories based on the Age Group column: Teen, Middle Age, Elder.

For example, it groups all teens together, all middle-aged customers together, and so on.

['Annual Income (k$)']:

From each group, we're selecting the numeric column Annual Income (k$) for analysis.

.agg(['mean', 'median', 'min', 'max', 'std']):

Applies multiple aggregation functions: average (mean), middle value (median), minimum, maximum, and standard deviation (std) for each group.

```python
# Group by 'Age Group' and compute summary statistics for 'Annual Income (k$)'
income_stats = df.groupby('Age Group')['Annual Income (k$)'].agg(
    ['mean', 'median', 'min', 'max', 'std']
)

# Display the results
income_stats
```

##  Create a List of Numeric Values for Each Categorical Variable
That means: for each Age Group, we should extract all numeric values of, say, Annual Income (k$) into a list.
Groups the dataset by Age Group.

For each group, gathers all Annual Income (k$) values into a Python list.

```python
# Create a dictionary of lists of income values grouped by Age Group
income_lists = df.groupby('Age Group')['Annual Income (k$)'].apply(list)

# Display the result
print(income_lists)
```
