```python
df1=pd.get_dummies(df1, columns=['part_time_job'], drop_first=True)
df1
```

```python
df1['part_time_job_Yes'] = df1['part_time_job_Yes'].astype(int)
df1
```

```python
#ps2type conersion
df1['diet_quality'] = df1['diet_quality'].astype('category')
df1['age']=df1['age'].astype('int')
df1
```

```python
#ps2check updated data type
df1.info()
```

```python
#ps2MinMax normalization
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df1[['study_hours_per_day', 'netflix_hours']] = scaler.fit_transform(df1[['study_hours_per_day', 'netflix_hours']])
df1
```

```python
#ps2 simple feature scaling
df1['study_hours_per_day'] = df1['study_hours_per_day'] / df1['study_hours_per_day'].max()
df1['netflix_hours'] = df1['netflix_hours'] / df1['netflix_hours'].max()
```

```python
#ps2 zscore
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df1[['study_hours_per_day', 'netflix_hours']] = scaler.fit_transform(df1[['study_hours_per_day', 'netflix_hours']])
```
