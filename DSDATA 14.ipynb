```python
df=pd.read_csv('/kaggle/input/weatherhistory/weatherHistory.csv')
df.head()
```

```python
df.info()
```

```python
# Check for missing/null values
print(df.isnull().sum())
```

```python
df['Precip Type'] = df['Precip Type'].fillna('unknown')
df
```

```python
#Convert Date to Datetime Format
df['Formatted Date'] = pd.to_datetime(df['Formatted Date'], utc=True)
df
```

```python
#Add Derived Time Columns i.e. extract the year and month
df['Year'] = df['Formatted Date'].dt.year
df['Month'] = df['Formatted Date'].dt.month
df
```

```python
df.info()
```

```python
yearly_avg = df.groupby('Year')[['Temperature (C)', 'Apparent Temperature (C)', 'Wind Speed (km/h)']].mean()
print(yearly_avg)
```

```python
monthly_avg = df.groupby('Month')[['Temperature (C)', 'Apparent Temperature (C)', 'Wind Speed (km/h)']].mean()
print(monthly_avg)
```

```python
import pandas as pd

# Ensure datetime is parsed correctly and timezone removed
df['Formatted Date'] = pd.to_datetime(df['Formatted Date'], utc=True).dt.tz_convert(None)

# Extract month name
df['Month'] = df['Formatted Date'].dt.strftime('%B')  # e.g., 'April'

# Group by month and calculate average temperature
avg_temp_per_month = df.groupby('Month')['Temperature (C)'].mean().reset_index()

# Optional: Sort by calendar month order
month_order = ['January', 'February', 'March', 'April', 'May', 'June',
               'July', 'August', 'September', 'October', 'November', 'December']
avg_temp_per_month['Month'] = pd.Categorical(avg_temp_per_month['Month'], categories=month_order, ordered=True)
avg_temp_per_month = avg_temp_per_month.sort_values('Month')

# Display result
print(avg_temp_per_month)
```
