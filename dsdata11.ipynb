## median is a better choice for skewed distributions

```python
# Fill missing 'age' values with the median
titanic['age'] = titanic['age'].fillna(titanic['age'].median())
```

## A boxplot that shows how ages are spread for males and females based on survival.

A violinplot that shows the same but in a more shapely and visual way.



## You can clearly see if survivors were generally younger or older, or if there was a difference between men and women.

```python
import matplotlib.pyplot as plt

# Step 3: Plot Boxplot for 'age' distribution by 'sex' and 'survived'
plt.figure(figsize=(14, 6))

# Boxplot
plt.subplot(1, 2, 1)
sns.boxplot(data=titanic, x='sex', y='age', hue='survived')
plt.title('Boxplot of Age by Sex and Survival Status')

# Violinplot
plt.subplot(1, 2, 2)
sns.violinplot(data=titanic, x='sex', y='age', hue='survived', split=True)
plt.title('Violinplot of Age by Sex and Survival Status')

plt.tight_layout()
plt.show()
```
